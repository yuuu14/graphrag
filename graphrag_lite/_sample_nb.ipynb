```python
from graphrag_lite.language_core.qwen_chat_llm import QwenChatLLM
from graphrag_lite.config.default import default_qwen_llm_config
from IPython.display import display, Markdown, update_display

from langchain_core.messages.human import HumanMessage
from langchain_core.messages.system import SystemMessage

print(default_qwen_llm_config)
llm = QwenChatLLM(config=default_qwen_llm_config)
print(llm._llm_string)

# question = "地球的质量？公式使用markdown $equation$的格式"
prompt = "Role: 你是一个写作大师",
task = "写作一篇200字关于月亮的散文",
messages = [
    SystemMessage(content=prompt),
    HumanMessage(content=task),
]
answer = ""
display_id = None
for part in llm.stream(messages):
    answer += part
    print(part, end="| ")
print()
```
